```python
import os

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import contextily as ctx
```

```python
plt.rcParams['keymap.quit'] = ['q', 'escape']
```

```python

# list all files in the directory
path = "data/Mobilfunkdaten/"

gdf = gpd.GeoDataFrame()

for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith(".geojson"):
            current_path = os.path.join(root, file)
            
            if not "Aussenantennen" in current_path:
                continue
                
            print(current_path)
            
            temp = gpd.read_file(os.path.join(root, file))
            
            gdf = pd.concat([gdf, temp], ignore_index=True)
gdf
```

```python
# plot the geodataframe (with rssi as color)
ax = gdf.plot(column='rssi', legend=True)

# add a basemap
ctx.add_basemap(ax,
                # url=ctx.providers.OpenStreetMap.Mapnik,
                crs=gdf.crs
                )
plt.show()
```
